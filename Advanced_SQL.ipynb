1. What is a Common Table Expression (CTE), and how does it improve SQL query readability?

-  A Common Table Expression (CTE) is a temporary named result set that is defined using the WITH clause. It exists only during the execution of a single SQL statement.

Advantages:

Makes complex queries easier to read and understand.
Breaks large queries into smaller, logical parts.
Avoids repeating the same subquery multiple times.
Supports recursive queries for hierarchical data.

WITH HighScorers AS (
    SELECT name, score
    FROM Student_Performance
    WHERE score > 85
)
SELECT *
FROM HighScorers;

2. Why are some views updatable while others are read-only? Explain with an example.

-  A view is a virtual table created from one or more tables.

Updatable views are based on a single table and do not contain complex SQL features like GROUP BY, DISTINCT, aggregate functions, or joins.
Read-only views contain joins, aggregate functions, GROUP BY, UNION, or DISTINCT, making it impossible for the database to determine how updates should be applied.

CREATE VIEW StudentView AS
SELECT student_id, name, score
FROM Student_Performance;

You can update this view:

UPDATE StudentView
SET score = 90
WHERE student_id = 101;

Example of a read-only view:

CREATE VIEW AvgScore AS
SELECT course, AVG(score) AS avg_score
FROM Student_Performance
GROUP BY course;

3. What advantages do stored procedures offer compared to writing raw SQL queries repeatedly?

-  A stored procedure is a precompiled collection of SQL statements stored in the database.

Advantages:

Improves performance because it is compiled once and reused.
Reduces code duplication.
Makes maintenance easier since changes are made in one place.
Enhances security by allowing users to execute procedures without direct table access.
Supports parameters, making procedures reusable for different inputs.

4. What is the purpose of triggers in a database? Mention one use case where a trigger is essential.

-  A trigger is a special stored program that automatically executes when a specific database event (INSERT, UPDATE, or DELETE) occurs.

Purpose:

Enforces business rules.
Maintains data integrity.
Automates tasks without user intervention.
Keeps audit logs of database changes.

Use Case:
A banking system can use a trigger to automatically record every transaction in an audit table whenever a customer's account balance is updated.

CREATE TRIGGER log_update
AFTER UPDATE ON Student_Performance
FOR EACH ROW
INSERT INTO Audit_Log(student_id, updated_on)
VALUES (NEW.student_id, NOW());

5. Explain the need for data modelling and normalization when designing a database.

-  Data Modelling

Data modelling is the process of designing the structure of a database by identifying entities, attributes, and relationships.

Benefits:

Organizes data efficiently.
Improves database design.
Reduces errors.
Makes future expansion easier.
Normalization

Normalization is the process of organizing data into multiple related tables to reduce redundancy and improve consistency.

Benefits:

Eliminates duplicate data.
Maintains data integrity.
Prevents insertion, update, and deletion anomalies.
Makes the database easier to maintain.

Category, TotalProducts, AveragePrice.

-  CREATE VIEW vw_CategorySummary AS
SELECT
    Category,
    COUNT(*) AS TotalProducts,
    AVG(Price) AS AveragePrice
FROM Products
GROUP BY Category;

To view the data:

SELECT * FROM vw_CategorySummary;

7.   Create an updatable view containing ProductID, ProductName, and Price.
Then update the price of ProductID = 1 using the view

-  To create an updatable view and then update the price of ProductID = 1, use the following SQL:

Step 1: Create the Updatable View

CREATE VIEW vw_ProductDetails AS
SELECT
    ProductID,
    ProductName,
    Price
FROM Products;

Step 2: Update the Price Using the View

For example, to change the price to 500:

UPDATE vw_ProductDetails
SET Price = 500
WHERE ProductID = 1;

Step 3: Verify the Update

SELECT * FROM vw_ProductDetails
WHERE ProductID = 1;
